# Creating New Columns

To analyse the data in more deeper way, lets few new columns: Total marks, Percentage and Grades.

```python
marks['Total_marks']=marks['Maths']+marks['Physics']+marks['Chemistry']+marks['Biology']+marks['History']+marks['Economics']+marks['Civics']+marks['English']
```

```python
marks['Total_marks']
```

```python
marks['Percentage']=marks['Total_marks']/800*100
```

```python
marks['Percentage']
```

```python
marks
```

Now lets create Grades

```python
#Assigning the grades

def determine_grade(scores):
    if scores >= 85 and scores <= 100:
        return 'Grade A'
    elif scores >= 70 and scores < 85:
        return 'Grade B'
    elif scores >= 55 and scores < 70:
        return 'Grade C'
    elif scores >= 35 and scores < 55:
        return 'Grade D'
    elif scores >= 0 and scores < 35:
        return 'Grade E'
    
marks['Grades']=marks['Percentage'].apply(determine_grade)
```

```python
marks.info()
```

##### Observations:

from above you may see that new column Grades had been added

```python
marks['Grades'].value_counts().plot.pie(autopct="%2.2f%%")
plt.show()
```

Most of the students achieve Grade C

```python
plt.title('Grades for each student',fontsize=12)
sns.countplot(x="Name", hue="Grades", data=marks)
plt.show()
```
